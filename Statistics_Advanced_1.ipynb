Question 6: What is the Central Limit Theorem (CLT), and why is it critical in statistics?
  - The Central Limit Theorem (CLT) states that :             
    - When you take many random samples form any population (no matter the original shape), the sampling distribution of the sample mean will become approximately normal (bell-shaped) as the sample size becomes large (usually n >= 30).
    - NO matter how the population looks, the distribution of sample means becomes NORMAL if the sample size is large.
  - Key Points of CLT
    1. Works for any population distribution
      - Skewed
      - Uniform
      - Bimodal
      - Not normal (sample means still become normal )  
    2. Must have independent and random samples.
    3. Works best when sample sie >= 30.
    4. The mean of the sample means = population mean.
  - Why is the CLT Critical in Statistics
    - The CLT is one of the most important concepts in all of statistics because :        
      - 1. Allows us to use the normal distribution
        - Evven if the data is not normal the CLT lets us use :       
          - Z-scores
          - Z-tables
          - Normal probability calculations   
      - 2. Foundation of confidence intervals
         - Every confidence interval formula (95 % CI) depends on the CLT.
      - 3. Basis of hypothesis testing
        - Test like
          - Z-test
          - t-test
          - Proportion test
          - all rely on CLT because sample means become normal.
      - 4. Used in data science and machine learning
        - CLT ensures that:
          - average behave predictably
          - model residuals often become normal
          - sample-based predictions are reliable
     - 5. Makes sampling practical
       - We already know the entire population.
       - CLT lets us:     
         - take a sample
         - compute its mean
         - generalize to the whole population

Question 9: Write a Python program to generate 1000 random numbers from a normal distribution with mean = 50 and standard deviation=5.     
Compute its mean and standard deviation using NumPy, and draw a histogram to visualize the distribution.

```python
import numpy as np
import matplotlib.pyplot as plt

def main():
    # Parameters for the normal distribution
    mu = 50      # Mean
    sigma = 5    # Standard deviation
    sample_size = 1000

    # 1. Generate 1000 random numbers from the normal distribution
    # np.random.normal takes loc (mean), scale (std dev), and size
    data = np.random.normal(loc=mu, scale=sigma, size=sample_size)

    # 2. Compute mean and standard deviation using NumPy
    computed_mean = np.mean(data)
    computed_std = np.std(data)

    # Print the results
    print(f"Target Mean: {mu}")
    print(f"Target Standard Deviation: {sigma}")
    print("-" * 30)
    print(f"Computed Mean: {computed_mean:.4f}")
    print(f"Computed Standard Deviation: {computed_std:.4f}")

    # 3. Draw a histogram to visualize the distribution
    plt.figure(figsize=(10, 6))

    # Plot histogram
    # 'bins' determines the number of bars, 'density=True' normalizes the plot
    count, bins, ignored = plt.hist(data, bins=30, density=True, alpha=0.6, color='b', edgecolor='black', label='Sample Data')

    #  Plot the theoretical probability density function (PDF) curve for comparison
    pdf_curve = (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((bins - mu) / sigma)**2)
    plt.plot(bins, pdf_curve, linewidth=2, color='r', label='Theoretical PDF')

    # Add labels and title
    plt.title('Histogram of 1000 Random Numbers (Normal Distribution)')
    plt.xlabel('Value')
    plt.ylabel('Density')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.5)

    # Show the plot
    plt.show()

if __name__ == "__main__":
    main()
```

Question 10: You are working as a data analyst for a retail company. The company has
collected daily sales data for 2 years and wants you to identify the overall sales trend.
daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
 235, 260, 245, 250, 225, 270, 265, 255, 250, 260]     
1. Explain how you would apply the Central Limit Theorem to estimate the average sales
with a 95% confidence interval.     
2. Write the Python code to compute the mean sales and its confidence interval.


Answer 10.1. To estimate the average sales with a 95% confidence interval using the provided sample:
  -  1. Calculate Statistics: Compute the sample mean ($\bar{x}$) and the sample standard deviation ($s$) of the 20 data points.
  - 2. Standard Error: Calculate the Standard Error of the Mean (SEM), which quantifies how much the sample mean is expected to vary from the true population mean. The formula is $SEM = \frac{s}{\sqrt{n}}$.
  - 3.  Critical Value: Determine the critical value for a 95% confidence level.Note: Strictly speaking, because the sample size ($n=20$) is small ($<30$) and the population standard deviation is unknown, we use the t-distribution rather than the normal (Z) distribution. If $n$ were large, we would use the Z-score (1.96). I will use the t-distribution in the code for statistical accuracy.
  - 4. Interval: The confidence interval is calculated as: $\bar{x} \pm (\text{Critical Value} \times SEM)$.

```python
# Answer 10.2
import numpy as np
import scipy.stats as stats

def main():
    # Daily sales data provided in the question
    daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
                   235, 260, 245, 250, 225, 270, 265, 255, 250, 260]

    # 1. Compute basic statistics
    n = len(daily_sales)
    mean_sales = np.mean(daily_sales)

    # Calculate sample standard deviation (ddof=1 for unbiased estimator)
    std_dev_sales = np.std(daily_sales, ddof=1)

    # 2. Compute Standard Error of the Mean (SEM)
    # Formula: s / sqrt(n)
    sem = std_dev_sales / np.sqrt(n)

    # 3. Determine the 95% Confidence Interval
    confidence_level = 0.95

    # Since n = 20 (which is < 30), we use the t-distribution for better accuracy.
    # Degrees of freedom
    degrees_of_freedom = n - 1

    # Calculate the interval using scipy.stats
    # loc is the center (mean), scale is the spread (standard error)
    confidence_interval = stats.t.interval(confidence_level, degrees_of_freedom, loc=mean_sales, scale=sem)

    # Output the results
    print(f"Data Analysis for Sales Trends")
    print("-" * 30)
    print(f"Sample Size (n): {n}")
    print(f"Mean Sales: {mean_sales:.2f}")
    print(f"Sample Standard Deviation: {std_dev_sales:.2f}")
    print(f"Standard Error (SEM): {sem:.2f}")
    print("-" * 30)
    print(f"95% Confidence Interval: ({confidence_interval[0]:.2f}, {confidence_interval[1]:.2f})")
    print("-" * 30)
    print(f"Interpretation: We are 95% confident that the true average daily sales \nfor the 2-year period lies between {confidence_interval[0]:.2f} and {confidence_interval[1]:.2f}.")

if __name__ == "__main__":
    main()
```
